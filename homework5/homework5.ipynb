```python
# 绘制直方图
def draw_hist(data, label, num_bins=10):
    plt.hist(data[label], bins=num_bins, color='green', alpha=0.8, edgecolor='white')

    # 添加标题和轴标签
    plt.grid()
    plt.title(f'Histogram of {label}')
    plt.xlabel(f'{label}')
    plt.ylabel('Frequency')

    # 显示图形
    plt.show()
```

接下来需要将评论量对数进行分组，并将原数据中其他符号类型变量转化为因子变量，由于转化为因子变量后会覆盖掉原始值，所以我这里先进行第三问的描述分析

```python
# 分析视频长度对因变量的影响，因为是连续变量对定性变量，所以绘制箱线图
data.boxplot(column='length_log', by='hot_or_not', figsize=(8, 6), fontsize='large')
plt.show()
```

```python
def draw_cross_tab(data, x_label, y_label):
    # 计算频数表
    freq_table = pd.crosstab(data[x_label], data[y_label])

    # 绘制堆叠条形图
    freq_table.plot(kind='bar', figsize=(8, 6), stacked=True, fontsize='large')

    # 添加标题和轴标签
    plt.grid()
    plt.title(f'Crosstab of {x_label}')

    # 显示图形
    plt.show()
```

```python
# 绘制话题类型对因变量的影响，因为是定性变量对定性变量，所以绘制堆叠柱状图
draw_cross_tab(data, 'type_sv', 'hot_or_not')
```

```python
# 绘制叙事角度对因变量的影响，因为是定性变量对定性变量，所以绘制堆叠柱状图
draw_cross_tab(data, 'nar_angle', 'hot_or_not')
```

```python
# 绘制态度倾向对因变量的影响，因为是定性变量对定性变量，所以绘制堆叠柱状图
draw_cross_tab(data, 'zf_sv', 'hot_or_not')
```

分析如下：
* 从视频长度对数箱线图看，热议视频的对数视频长度更小，而且箱体更窄，可能短视频更能引起热议，太长的视频会让用户失去看的耐心
* 从话题类型堆叠柱状图来看，传统话题和非传统话题的热议视频占比和非热议视频占比差不多，被动回应这一话题类型中热议视频占比较小，另外传统话题和非传统话题的视频数量比较大，可能用户整体对被动回应类型的视频不太感兴趣
* 从叙事角度堆叠柱状图来看，明星团队这类视频中热议视频占比较高，可能用户对娱乐圈新闻更感兴趣，品牌宣传的热议视频占比较低，可能这类属于广告，所以用户不感兴趣，赛事周边和要素解读这两类中热议视频和非热议视频占比相差不大
* 从态度倾向堆叠柱状图来看，态度倾向为“中”和“正”的视频中热议视频和非热议视频占比相差不大，态度为“负”的视频中热议视频占比相对较低，可能负面内容的曝光率相对较低，或者用户更喜欢正面一些的内容

接下来进行因子变量转换，划分训练集和测试集，构造模型

```python
# 需要将定性变量转变为因子变量 y_label也需要转换
discrete_label = ['type_sv', 'nar_angle', 'zf_sv', 'media_type', 'type_time_effe']

continuous_label = ['length_log', 'media_fans_log', 'love_sv_log']

# 全部的自变量label
x_label = ['length_log', 'media_fans_log', 'love_sv_log', 'type_sv', 'nar_angle', 'zf_sv', 'media_type', 'type_time_effe', 'month1']

# y label
y_label = 'hot_or_not'
```

```python
converted_data = copy.deepcopy(data)
converted_data[y_label] = np.where(data['com_sv_log'] >= 2, 1, 0)

for cur_label in discrete_label:
    converted_data[cur_label] = pd.factorize(data[cur_label])[0]

# 对连续型变量进行标准化处理
scaler = StandardScaler()
converted_data[continuous_label] = scaler.fit_transform(data[continuous_label])
```

```python
# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(converted_data[x_label], converted_data[y_label], 
                                                    test_size=0.2, random_state=2023)
```

```python
# 构建决策树分类器
search_clf = DecisionTreeClassifier(random_state=2023)

# 定义参数搜索范围
param_grid = {'criterion': ['gini', 'entropy'], 'max_depth': [4, 6, 8, 10, 12], 
              'min_samples_split': [2, 4, 6, 8], 'min_samples_leaf': [1, 2]}

# 使用GridSearchCV进行五折交叉验证，并对参数进行调优
# 需要考虑如何划分k折，因此也使用一个固定的随机数种子2023
kf = KFold(n_splits=5, shuffle=True, random_state=2023)
grid_search_decision_tree = GridSearchCV(search_clf, param_grid=param_grid, cv=5, scoring='roc_auc')
grid_search_decision_tree.fit(X_train, y_train)

# 输出最佳参数组合
print("Best parameters: ", grid_search_decision_tree.best_params_)

# 在测试集上评估模型性能
print("Test set accuracy: ", accuracy_score(y_test, grid_search_decision_tree.predict(X_test)))
```

```python
# 构建AdaBoost分类器，需要为base_estimator设置随机数种子
ada_clf = AdaBoostClassifier(random_state=2023)

param_grid = {'n_estimators': [50, 100, 200],
              'learning_rate': [0.01, 0.1, 1]}

kf = KFold(n_splits=5, shuffle=True, random_state=2023)
grid_search_ada = GridSearchCV(ada_clf, param_grid=param_grid, cv=5, scoring='roc_auc')
grid_search_ada.fit(X_train, y_train)

# 输出最佳参数组合
print("Best parameters: ", grid_search_ada.best_params_)

# 在测试集上评估模型性能
print("Test set accuracy: ", accuracy_score(y_test, grid_search_ada.predict(X_test)))
```

```python
# 构建RandomForest分类器
forest_clf = RandomForestClassifier(random_state=2023)

param_grid = {'n_estimators': [10, 50, 100],
              'max_depth': [4, 6, 8, 10],
              'min_samples_split': [2, 4, 6, 8],
              'max_features': ['sqrt', 'log2', None]}

kf = KFold(n_splits=5, shuffle=True, random_state=2023)
grid_search_forest = GridSearchCV(forest_clf, param_grid=param_grid, cv=5, scoring='roc_auc')
grid_search_forest.fit(X_train, y_train)

# 输出最佳参数组合
print("Best parameters: ", grid_search_forest.best_params_)

# 在测试集上评估模型性能
print("Test set accuracy: ", accuracy_score(y_test, grid_search_forest.predict(X_test)))
```

```python
def draw_roc(model_name, y_pred, y_test):
    # 计算FPR, TPR和阈值
    fpr, tpr, thresholds = roc_curve(y_test, y_pred)

    # 计算AUC值
    auc_score = roc_auc_score(y_test, y_pred)

    # 绘制ROC曲线
    plt.plot(fpr, tpr, label='Auc: %0.4f)' % auc_score)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])  # 将y轴范围扩展一些，使得ROC曲线的最高点能够完整地展示出来
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} ROC')
    plt.legend(loc="lower right")
    plt.show()
```

```python
draw_roc('DecisionTree', y_test, grid_search_decision_tree.predict(X_test))
```

```python
draw_roc('AdaBoost', y_test, grid_search_ada.predict(X_test))
```

```python
draw_roc('RandomForest', y_test, grid_search_forest.predict(X_test))
```

从测试集结果来看，随机森林的性能最好，auc为0.9291，AdaBoost的性能次之，auc为0.9206，决策树的性能在三者中最低，auc为0.8942。总体上三个模型的性能都比较不错，基本上达到了auc 0.9的水平
