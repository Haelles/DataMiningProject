# 2 基于商品名称的自动化分类

## 2.1 数据读入

读入数据集 catalogs.csv，该数据包含每个商品的名称以及对应的一二三级分类，并查看基本形式。

```python
import numpy as np
import pandas as pd
import jieba
import re
from collections import Counter
from wordcloud import WordCloud
from functools import reduce
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

import warnings
warnings.filterwarnings('ignore')
```

```python
catalog = pd.read_csv('./data/catalogs.csv')
catalog.head()
```

## 2.2 一级品类下的二级品类分布饼状图

对每个一级品类下的各个二级品类的分布情况用饼状图进行描述分析。

```python
# 生鲜一级品类下各个二级品类的分布情况
fresh = catalog[catalog['first'] == '生鲜']
fresh_count = fresh['second'].value_counts()
fresh_labels = list(fresh_count.index)

# 食品饮料、保健食品一级品类下各个二级品类的分布情况
food = catalog[catalog['first'] == '食品饮料、保健食品']
food_count = food['second'].value_counts()
food_labels = list(food_count.index)

# 饼状图绘制
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.pie(fresh_count, labels=fresh_labels, autopct='%1.1f%%')
ax1.set_title('生鲜一级品类下的二级品类分布')
ax2.pie(food_count, labels=food_labels, autopct='%1.1f%%')
ax2.set_title('食品饮料、保健食品一级品类下的二级品类分布')
plt.show()
```

文字解读

## 2.3 分词

利用分词工具对一级品类为“生鲜”的数据进行分词处理，并做出来分词后的词云图。

```python
# 加载自定义词典
jieba.load_userdict('./data/userdict.dat')

# 对一级品类为“生鲜”的数据进行分词处理
fresh = catalog[catalog['first'] == '生鲜']
fresh_name = fresh['name'].tolist()
fresh_seg_list = []
for name in fresh_name:
    seg_list = jieba.lcut(str(name))
    fresh_seg_list.append(seg_list)

fresh['seg_list'] = fresh_seg_list
fresh.head()
```

```python
# 将多个列表中的词汇合并为一个列表
all_words = reduce(lambda x, y: x+y, fresh['seg_list'])

# 将列表中的词汇用空格连接成一个长字符串
text = ' '.join(all_words)

# 生成词云
wc = WordCloud(background_color='white', max_words=200, font_path='simsun.ttf')

plt.imshow(wc.generate(text))
plt.axis('off')
plt.show()
```

## 2.4 分词结果建模预处理

对分词结果进行预处理：去掉数字字母组合，然后去掉低频词（包含该词的商品数量<10）、高频词（在多于75%的商品名称中出现）。

```python
# 去掉数字字母组合
pattern = re.compile(r'\w*\d\w*')
fresh['seg_list'] = [[word for word in seg_list if not pattern.match(word)] for seg_list in fresh['seg_list']]

# 去除符号
def clean_text(text_list):
    res = []
    for text in text_list:
        res.append(re.sub(r"[{}]+".format('.,;《》？！“”‘’@#￥%…&×（）——+【】{};；●，。&～、|\s:：' + '()丨/'), "", text))
    return res
fresh['seg_list'] = fresh['seg_list'].apply(clean_text)


# 去除停用词
stopwords = pd.read_csv('cn_stopwords.txt', encoding='utf8', names=['stopword'], index_col=False)
stop_list = stopwords['stopword'].tolist()
fresh['seg_list'] = [[word for word in seg_list if word not in stop_list] for seg_list in fresh['seg_list']]

# 统计每个词出现的次数
word_counts = Counter([word for seg_list in fresh['seg_list'] for word in seg_list])

# 去掉低频词和高频词
low_frequency_words = set([word for word in word_counts if word_counts[word] < 10])
high_frequency_words = set([word for word in word_counts if word_counts[word]/len(fresh) > 0.75])
fresh['seg_list'] = [[word for word in seg_list if word not in low_frequency_words and word not in high_frequency_words] for seg_list in fresh['seg_list']]

# 将分词结果转换为字符串格式
fresh['content'] = fresh['seg_list'].apply(lambda x: ' '.join(x))

fresh.head()
```

## 2.5 朴素贝叶斯模型

将上一步得到的预处理后的数据按照 8:2 的方式划分训练集和测试集。保留高频词中的 top1000 的词语，作为分类器的特征，并建立文档-词频矩阵。

```python
# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(fresh['content'], fresh['second'], test_size=0.2, random_state=42)

# 提取高频词中的top1000作为分类器的特征
word_counts = Counter([word for seg_list in fresh['seg_list'] for word in seg_list])
top1000 = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)[:1000]
vocab = [pair[0] for pair in top1000]

train_counts = CountVectorizer(vocabulary=vocab).fit_transform(X_train)
print(f"训练集文档-词频矩阵的维度为：{train_counts.shape}")
test_counts = CountVectorizer(vocabulary=vocab).fit_transform(X_test)
print(f"测试集文档-词频矩阵的维度为：{test_counts.shape}")
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

clf = MultinomialNB()
clf.fit(train_counts, y_train)

y_pred = clf.predict(test_counts)
accuracy = accuracy_score(y_test, y_pred)
print(f"在测试集上的准确率为：{accuracy}")
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)

labels = sorted(fresh['second'].unique())
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', xticklabels=labels, yticklabels=labels, cmap='Blues')
ax.set_xlabel('预测标签')
ax.set_ylabel('真实标签')
ax.set_title('混淆矩阵')
plt.show()
```
